```python
import pandas as pd
df=pd.read_csv("/content/diabetes.csv")
df.head()
```

```python
df["Outcome"].value_counts()#imbalanced dataset
```

```python
x=df.iloc[:,:-1]
x
y=df.iloc[:,-1]
y
```

```python
#oversampling-268-500

from imblearn.over_sampling import SMOTE

sm=SMOTE(random_state=42)#random_state-to keep new samples constant
x_new,y_new=sm.fit_resample(x,y)#store new x and y values
```

```python
#to check oversampling
from collections import Counter#Counter-to get count of samples
print(Counter(y))
print(Counter(y_new))
```

```python
from imblearn.under_sampling import RandomUnderSampler #undersampling
ru=RandomUnderSampler(random_state=42)
x_ud,y_ud=ru.fit_resample(x,y)#fit_resample-create new samples
```

```python
from collections import Counter
print(Counter(y))
print(Counter(y_ud))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

x_train_os,x_test_os,y_train_os,y_test_os=train_test_split(x_new,y_new,test_size=0.2,random_state=42)

x_train_ud,x_test_ud,y_train_ud,y_test_ud=train_test_split(x_ud,y_ud,test_size=0.2,random_state=42)
```

```python
#normal data
from sklearn.svm import SVC
from sklearn.metrics import  accuracy_score
sv=SVC(kernel='linear')
sv.fit(x_train,y_train)
y_predict=sv.predict(x_test)
print(accuracy_score(y_test,y_predict))
```

```python
#oversampled data
sv.fit(x_train_os,y_train_os)
y_predict_os=sv.predict(x_test_os)
print(accuracy_score(y_test_os,y_predict_os))
```

```python
#undersampled data
sv.fit(x_train_ud,y_train_ud)
y_predict_ud=sv.predict(x_test_ud)
print(accuracy_score(y_test_ud,y_predict_ud))
```

```python
#naive bayes
#normal data
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
gnb=GaussianNB()
gnb.fit(x_train,y_train)
y_predict=gnb.predict(x_test)
print(accuracy_score(y_test,y_predict))
```

```python
gnb.fit(x_train_ud,y_train_ud)#undersampled
y_predict_ud =gnb.predict(x_test_ud)
print(accuracy_score(y_test_ud, y_predict_ud))
```

```python
gnb.fit(x_train_os,y_train_os)#undersampled
y_predict_os =gnb.predict(x_test_os)
print(accuracy_score(y_test_os, y_predict_os))
```
